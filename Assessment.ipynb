1. a)-----We will use the so--called Gillespie algorithm to simulate the 2-stage model for gene expression. 

```python
%pylab inlineimport numpy as np
```

We will start by defining the functions that are the conditional intensisties for each individual reaction. Transcription is reaction 0, translation is reaction 1mRNA degradationi reaction 2 and finally protein degradation is reaction 3.

```python
def lamb_0(X):    return 1E-2def lamb_1(X):        M = X[0]        return M*5E-3def lamb_2(X):        M = X[0]        return M*1E-3def lamb_3(X):        Z = X[1]        return Z*1E-4
```

To sample the time we must wait for the next reaction to occur we simply sample from an exponential distributionwhose mean is the inverse of the sum of the conditional intensities for the 4 reactions:

```python
def sample_wait_time(X, sum_of_rates):        M, Z = X        scale =  sum_of_rates        wait_time = np.random.exponential(scale)        return wait_time
```

To *choose* which reaction happens we sample them from a distribution where the probability for each reaction to occur is proportional to its conditional intensity.Explicitly:$$\mathrm{Pr}[\mathrm{reaction}=j] = \frac{\lambda_j(X^{R[j]}_i)}{ \sum_m \lambda_m(X^{R[m]}_i)}$$To do this we generate a uniform random number in the interval $[0,1]$ and split the interval in smaller sizes corresponding to the probability of that reaction ocurring.

```python
def choose_reaction_type(rates_list, sum_of_rates):        P = [(rate/sum_of_rates) for rate in rates_list]        random_number = np.random.uniform()        for i in range(4):        if random_number <= sum(P[0:i+1]):            reaction_type = i            return reaction_type        else:            continue        print "error in choosing reaction type"
```

Finally we just define a function which will carry out the steps of the algorithm calling these other functions. The algorithm consists of sampling the time interval between reactionsthen choosing the reaction type to update $X$ an repeat.

```python
def two_stage_process(iterations, X0, S, rates_list):        X = X0.copy()    t = 0        T = []    X_list = []        T.append(t)    X_list.append(X)        for i in range(iterations):                rates_x = [rate(X) for rate in rates_list]        sum_of_rates = sum(rates_x)                t += sample_wait_time(X, sum_of_rates)        reaction = choose_reaction_type(rates_x, sum_of_rates)        X_new = X + S[:,reaction]                T.append(t)        X_list.append(X_new)                X = X_new            T = np.array(T)    X_list = np.array(X_list)        return T, X_list
```

We now set the rates and the stoichiometric matrix:

```python
rates_list = [lamb_0, lamb_1, lamb_2, lamb_3]S = np.array([1.,0.,-1.,0.,0.,1.,0.,-1.]).reshape(2,4)print "S = \n", S
```

```python
T, X_list = two_stage_process(10000, np.zeros(2), S, rates_list)plot(T, X_list)
```

From the graph above, in which 10000 reactions were allowed to happen, we can take 5000 reactions (for these values of the rates of the reactions) as a good *burn-in* period.It is probably safe to take less reactions than that, perhaps around 3000, however we will use 5000 in what remains of this exercise.

The transcription reaction can be written in the following way:$$\mathrm{gene} \rightarrow \mathrm{gene} + M$$with a reaction rate of $u$. In this case the intensity of the process is independent of the state of the system and is also constant throughout the realisation time.This fits the definition of a Poisson process. Therefore the counting process $B(t)$ that counts the numer of transcritpion events that have occurred up to time $t$is a Poisson process with parameter (intensity) $u$:$$B(t) \sim \mathrm{Poisson}(u).$$Which means that:$$\mathrm{Pr}[N(t + \tau) - N(t) =k] = \frac{e^{-u\tau} (u\tau)^{k}}{k!}.$$We also know that the $E[B(t)] = u$ since it is a property of Poisson processes. Therefore $E[B(t)/u] = \dfrac{1}{u} E[B(t)] = 1$
